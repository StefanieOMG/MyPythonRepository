```python
def plot_word_cloud(data, swords):
    text = ''.join(data)
    words = list(jieba.cut(text))
    ex_sw_words = []
    for word in words:
        if len(word)>1 and (word not in swords):
            ex_sw_words.append(word)
    c = Counter()
    c = Counter(ex_sw_words)
    wc_data = pd.DataFrame({'word':list(c.keys()), 'counts':list(c.values())}).sort_values(by='counts', ascending=False).head(100)
    wc_data = [(x, y) for x, y in wc_data.values]
    wordcloud = (WordCloud()
                .add('',wc_data, word_size_range=[20, 100])
                )
    wordcloud.render('wordcloud.html')
    return wordcloud
```

```python
plot_word_cloud(data=data_true.loc[-data_true['clean_text'].str.contains("è½¬å‘"), 'clean_text'], swords=swords)
```
